```python
corpus ='''
In today’s fast-paced digital world, technology has become an inseparable part of our daily lives. From smartphones and laptops to smartwatches and AI-powered assistants, modern innovations shape how we communicate, work, and think. As society becomes more interconnected, the demand for digital literacy continues to rise. Students are taught to code in schools, employees attend workshops to upskill themselves in the latest tools, and businesses constantly invest in digital transformation to stay ahead. This shift isn’t just about convenience — it’s about survival. Companies that fail to innovate often get left behind in a highly competitive environment. Startups that embrace new technologies like blockchain, machine learning, and big data analytics often disrupt traditional industries, sometimes changing them entirely. For example, ride-sharing apps revolutionized urban transportation, and streaming platforms have redefined how we consume entertainment. All of this is made possible by the seamless integration of software, hardware, and data science.

However, with great advancement comes great responsibility. The increased dependence on digital tools raises concerns about privacy, data security, and the ethical use of technology. Social media, once hailed as a revolutionary way to connect with friends and share life experiences, has also exposed users to risks like cyberbullying, misinformation, and identity theft. Tech companies are under constant scrutiny to maintain transparency and protect user information, but regulations often lag behind the pace of innovation. Governments around the world are now realizing the importance of setting up frameworks that balance technological progress with public welfare. Digital literacy doesn’t just mean knowing how to use devices — it also includes understanding the consequences of our online behavior, recognizing fake news, and being aware of how personal data is collected and used.
'''
```

```python
import nltk
```

```python
from nltk.tokenize import LineTokenizer
```

```python
line_tokenizer=LineTokenizer()
```

```python
sentences=line_tokenizer.tokenize(corpus)
```

```python
sentences
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences1=sent_tokenize(corpus)
```

```python
sentences1
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stopwords.words('french')
```

```python
ENG_STOPWORDS=stopwords.words('english')
```

```python
ENG_STOPWORDS
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
from nltk.tokenize import word_tokenize
```

```python
for i in range(len(sentences1)):
    words=nltk.word_tokenize(sentences1[i])
    words=[stemmer.stem(word.lower()) for word in words if word not in set(ENG_STOPWORDS)]
    sentences1[i]=' '.join(words)
```
